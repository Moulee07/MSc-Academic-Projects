```python
import pandas as pd

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import plotly.express as px
```

```python
Day_Barcelona_full_weather_data = pd.read_csv('Day_Barcelona_full_weather_data.csv')
Day_Bilbao_full_weather_data= pd.read_csv('Day_Bilbao_full_weather_data.csv')
Day_Madrid_full_weather_data= pd.read_csv('Day_Madrid_full_weather_data.csv')
Day_Seville_full_weather_data= pd.read_csv('Day_Seville_full_weather_data.csv')
Day_Valencia_full_weather_data= pd.read_csv('Day_Valencia_full_weather_data.csv')


daily_average_solar = pd.read_csv('daily_average_solar.csv')
```

```python
Day_Barcelona_full_weather_data['dt_iso'] = pd.to_datetime(Day_Barcelona_full_weather_data['dt_iso'],
                                                           utc=True, 
                                                           errors='raise')
Day_Bilbao_full_weather_data['dt_iso'] = pd.to_datetime(Day_Bilbao_full_weather_data['dt_iso'],
                                                           utc=True, 
                                                           errors='raise')
Day_Madrid_full_weather_data['dt_iso']  = pd.to_datetime(Day_Madrid_full_weather_data['dt_iso'],
                                                           utc=True, 
                                                           errors='raise')
Day_Seville_full_weather_data['dt_iso']  = pd.to_datetime(Day_Seville_full_weather_data['dt_iso'],
                                                           utc=True, 
                                                           errors='raise')
Day_Valencia_full_weather_data['dt_iso']  = pd.to_datetime(Day_Valencia_full_weather_data['dt_iso'],
                                                           utc=True, 
                                                           errors='raise')
```

```python
Day_Bilbao_full_weather_data.set_index('dt_iso', inplace=True)

# Resample by day and calculate daily statistics
daily_resample = Day_Bilbao_full_weather_data.resample('D')

# Calculate daily statistics for the new DataFrame
Bilbao_day_data = pd.DataFrame({
    'temp_min': daily_resample['temp'].min(),
    'temp_max': daily_resample['temp'].max(),
    'temp_avg': daily_resample['temp'].mean(),
    'pressure': daily_resample['pressure'].mean(),
    'wind_speed': daily_resample['wind_speed'].mean(),
    'humidity':daily_resample['humidity'].mean(),
    'rain': daily_resample['rain_1h'].sum(),  # Total rain for the day
    'snow': daily_resample['snow_3h'].sum(),  # Total snow for the day
    'cloud_cover': daily_resample['clouds_all'].mean()  # Average cloud cover
})

Bilbao_day_data = Bilbao_day_data.add_prefix('Bilbao_')
```

```python
Day_Barcelona_full_weather_data.set_index('dt_iso', inplace=True)

# Resample by day and calculate daily statistics
daily_resample = Day_Barcelona_full_weather_data.resample('D')

# Calculate daily statistics for the new DataFrame
Barcelona_day_data = pd.DataFrame({
    'temp_min': daily_resample['temp'].min(),
    'temp_max': daily_resample['temp'].max(),
    'temp_avg': daily_resample['temp'].mean(),
    'pressure': daily_resample['pressure'].mean(),
    'wind_speed': daily_resample['wind_speed'].mean(),
    'humidity':daily_resample['humidity'].mean(),
    'rain': daily_resample['rain_1h'].sum(),  # Total rain for the day
    'snow': daily_resample['snow_3h'].sum(),  # Total snow for the day
    'cloud_cover': daily_resample['clouds_all'].mean()  # Average cloud cover
})

Barcelona_day_data = Barcelona_day_data.add_prefix('Barcelona_')
```

```python
Day_Madrid_full_weather_data.set_index('dt_iso', inplace=True)

# Resample by day and calculate daily statistics
daily_resample = Day_Madrid_full_weather_data.resample('D')

# Calculate daily statistics for the new DataFrame
Madrid_day_data = pd.DataFrame({
    'temp_min': daily_resample['temp'].min(),
    'temp_max': daily_resample['temp'].max(),
    'temp_avg': daily_resample['temp'].mean(),
    'pressure': daily_resample['pressure'].mean(),
    'wind_speed': daily_resample['wind_speed'].mean(),
    'humidity':daily_resample['humidity'].mean(),
    'rain': daily_resample['rain_1h'].sum(),  # Total rain for the day
    'snow': daily_resample['snow_3h'].sum(),  # Total snow for the day
    'cloud_cover': daily_resample['clouds_all'].mean()  # Average cloud cover
})

Madrid_day_data = Madrid_day_data.add_prefix('Madrid_')
```

```python
Day_Seville_full_weather_data.set_index('dt_iso', inplace=True)

# Resample by day and calculate daily statistics
daily_resample = Day_Seville_full_weather_data.resample('D')

# Calculate daily statistics for the new DataFrame
Seville_day_data = pd.DataFrame({
    'temp_min': daily_resample['temp'].min(),
    'temp_max': daily_resample['temp'].max(),
    'temp_avg': daily_resample['temp'].mean(),
    'pressure': daily_resample['pressure'].mean(),
    'wind_speed': daily_resample['wind_speed'].mean(),
    'humidity':daily_resample['humidity'].mean(),
    'rain': daily_resample['rain_1h'].sum(),  # Total rain for the day
    'snow': daily_resample['snow_3h'].sum(),  # Total snow for the day
    'cloud_cover': daily_resample['clouds_all'].mean()  # Average cloud cover
})

Seville_day_data = Seville_day_data.add_prefix('Seville_')
```

```python
Day_Valencia_full_weather_data.set_index('dt_iso', inplace=True)

# Resample by day and calculate daily statistics
daily_resample = Day_Valencia_full_weather_data.resample('D')

# Calculate daily statistics for the new DataFrame
Valencia_day_data = pd.DataFrame({
    'temp_min': daily_resample['temp'].min(),
    'temp_max': daily_resample['temp'].max(),
    'temp_avg': daily_resample['temp'].mean(),
    'pressure': daily_resample['pressure'].mean(),
    'wind_speed': daily_resample['wind_speed'].mean(),
    'humidity':daily_resample['humidity'].mean(),
    'rain': daily_resample['rain_1h'].sum(),  # Total rain for the day
    'snow': daily_resample['snow_3h'].sum(),  # Total snow for the day
    'cloud_cover': daily_resample['clouds_all'].mean()  # Average cloud cover
})

Valencia_day_data = Valencia_day_data.add_prefix('Valencia_')
```

```python
# List of DataFrames to be merged horizontally
dataframes_to_merge = [Valencia_day_data, Seville_day_data, Madrid_day_data, Barcelona_day_data, Bilbao_day_data]

# Merge horizontally using concat with axis=1
day_weather_merged_df = pd.concat(dataframes_to_merge, axis=1)  # Axis 1 indicates horizontal merge

daily_average_solar['time'] = pd.to_datetime(daily_average_solar['time'], errors='raise')
```

```python
# Merge horizontally based on 'dt_iso' and 'time'
merged_df = pd.merge(
    day_weather_merged_df,
    daily_average_solar,
    left_index=True,  # Use 'dt_iso' as the key for the left DataFrame
    right_on='time',  # Use 'time' as the key for the right DataFrame
    how='inner'  # Use 'inner' to keep only matching rows
)
merged_df.set_index('time', inplace=True) 
```

```python
merged_df['time'] = pd.to_datetime(merged_df.index, errors='raise', utc=True)
# Create a new column to determine if it's a weekend or weekday
merged_df['day_of_week'] = merged_df['time'].dt.dayofweek  # 0=Monday, 6=Sunday

# Define whether it's a weekend or weekday
merged_df['is_weekend'] = merged_df['day_of_week'].apply(lambda x: 'Weekend' if x >= 5 else 'Weekday')

# Create a new column with the month extracted from 'dt_iso'
merged_df['month'] = merged_df['time'].dt.strftime('%b') 

# One-hot encode 'is_weekend'
merged_df = pd.get_dummies(merged_df, columns=['is_weekend'], prefix='is_weekend', drop_first=True)
```

```python
# Define the split point (end of training data)
split_point = '2017-12-31'  # Last date for training data

# Create training data (before the split point)
train_data = merged_df[merged_df['time'] <= split_point]

# Create testing data (after the split point)
test_data = merged_df[merged_df['time'] > split_point]

# Display the size of training and testing sets
print("Training Data Size:", train_data.shape)
print("Testing Data Size:", test_data.shape)
```

```python
# Features to use for training and testing
features = ['Valencia_temp_avg', 'Valencia_cloud_cover',
            'Seville_temp_avg', 'Seville_cloud_cover',
            'Madrid_temp_avg', 'Madrid_cloud_cover',
            'Barcelona_temp_avg', 'Barcelona_cloud_cover',
            'Bilbao_temp_avg', 'Bilbao_cloud_cover'
           ]

target = 'generation solar'
```

```python
# Train the Random Forest model on the training data
model = RandomForestRegressor(n_estimators=500, random_state=42)
model.fit(train_data[features], train_data[target])

# Predict on the testing set
y_pred = model.predict(test_data[features])


# Calculate Mean Squared Error to evaluate model performance
mse = mean_squared_error(test_data[target], y_pred)
print("Mean Squared Error:", mse)
```

```python
fig = px.scatter(x=test_data['time'], y=test_data[target], title='Actual vs. Predicted Solar Generation')
fig.add_scatter(x=test_data['time'], y=y_pred, mode='lines', name='Predicted')

# Update layout to set background color to white and display axes
fig.update_layout(
    xaxis=dict(
        title='Days',  # Set x-axis title
        showline=True,  # Display x-axis line
        linecolor='black',  # Set x-axis line color
        linewidth=2,  # Set x-axis line width
    ),
    yaxis=dict(
        title='Electricity Generation in GWh',  # Set y-axis title
        showline=True,  # Display y-axis line
        linecolor='black',  # Set y-axis line color
        linewidth=2,  # Set y-axis line width
    ),
    paper_bgcolor='white',  # Set plot background color to white
    plot_bgcolor='white',  # Set figure background color to white
)
fig.show()
```
