```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
Barcelona_full_weather_data = pd.read_csv('Barcelona_full_weather_data.csv')
Bilbao_full_weather_data= pd.read_csv('Bilbao_full_weather_data.csv')
Madrid_full_weather_data= pd.read_csv('Madrid_full_weather_data.csv')
Seville_full_weather_data= pd.read_csv('Seville_full_weather_data.csv')
Valencia_full_weather_data= pd.read_csv('Valencia_full_weather_data.csv')

# Spain Energy Data
energy_data = pd.read_csv('daily_avg_max_load.csv')
```

```python
Barcelona_full_weather_data = Barcelona_full_weather_data[['dt_iso','city_name','temp','pressure','humidity',
                                                           'wind_speed','rain_1h','snow_3h','clouds_all']]
Madrid_full_weather_data = Madrid_full_weather_data[['dt_iso','city_name','temp','pressure','humidity',
                                                           'wind_speed','rain_1h','snow_3h','clouds_all']]
Seville_full_weather_data = Seville_full_weather_data[['dt_iso','city_name','temp','pressure','humidity',
                                                           'wind_speed','rain_1h','snow_3h','clouds_all']]
Valencia_full_weather_data = Valencia_full_weather_data[['dt_iso','city_name','temp','pressure','humidity',
                                                           'wind_speed','rain_1h','snow_3h','clouds_all']]
Bilbao_full_weather_data = Bilbao_full_weather_data[['dt_iso','city_name','temp','pressure','humidity',
                                                           'wind_speed','rain_1h','snow_3h','clouds_all']]
```

```python
Bilbao_full_weather_data['dt_iso']= pd.to_datetime(Bilbao_full_weather_data['dt_iso'],
                                        utc=True, 
                                        errors='raise')
Valencia_full_weather_data['dt_iso']= pd.to_datetime(Valencia_full_weather_data['dt_iso'],
                                        utc=True, 
                                        errors='raise')
Barcelona_full_weather_data['dt_iso']= pd.to_datetime(Barcelona_full_weather_data['dt_iso'],
                                        utc=True, 
                                        errors='raise')
Seville_full_weather_data['dt_iso'] = pd.to_datetime(Seville_full_weather_data['dt_iso'],
                                        utc=True, 
                                        errors='raise')
Madrid_full_weather_data['dt_iso'] = pd.to_datetime(Madrid_full_weather_data['dt_iso'],
                                        utc=True, 
                                        errors='raise')
```

```python
# Converting Hourly to daily Data:


Bilbao_full_weather_data.set_index('dt_iso', inplace=True)

# Resample by day and calculate daily statistics
daily_resample = Bilbao_full_weather_data.resample('D')

# Calculate daily statistics for the new DataFrame
Bilbao_data = pd.DataFrame({
    'temp_min': daily_resample['temp'].min(),
    'temp_max': daily_resample['temp'].max(),
    'temp_avg': daily_resample['temp'].mean(),
    'pressure': daily_resample['pressure'].mean(),
    'wind_speed': daily_resample['wind_speed'].mean(),
    'humidity':daily_resample['humidity'].mean(),
    'rain': daily_resample['rain_1h'].sum(),  # Total rain for the day
    'snow': daily_resample['snow_3h'].sum(),  # Total snow for the day
    'cloud_cover': daily_resample['clouds_all'].mean()  # Average cloud cover
})

Bilbao_data = Bilbao_data.add_prefix('Bilbao_')
```

```python
# Converting Hourly to daily Data:

Valencia_full_weather_data.set_index('dt_iso', inplace=True)

# Resample by day and calculate daily statistics
daily_resample = Valencia_full_weather_data.resample('D')

# Calculate daily statistics for the new DataFrame
Valencia_data = pd.DataFrame({
    'temp_min': daily_resample['temp'].min(),
    'temp_max': daily_resample['temp'].max(),
    'temp_avg': daily_resample['temp'].mean(),
    'pressure': daily_resample['pressure'].mean(),
    'wind_speed': daily_resample['wind_speed'].mean(),
    'humidity':daily_resample['humidity'].mean(),
    'rain': daily_resample['rain_1h'].sum(),  # Total rain for the day
    'snow': daily_resample['snow_3h'].sum(),  # Total snow for the day
    'cloud_cover': daily_resample['clouds_all'].mean()  # Average cloud cover
})

Valencia_data = Valencia_data.add_prefix('Valencia_')
```

```python
# Converting Hourly to daily Data:
Barcelona_full_weather_data.set_index('dt_iso', inplace=True)

# Resample by day and calculate daily statistics
daily_resample = Barcelona_full_weather_data.resample('D')

# Calculate daily statistics for the new DataFrame
Barcelona_data = pd.DataFrame({
    'temp_min': daily_resample['temp'].min(),
    'temp_max': daily_resample['temp'].max(),
    'temp_avg': daily_resample['temp'].mean(),
    'pressure': daily_resample['pressure'].mean(),
    'wind_speed': daily_resample['wind_speed'].mean(),
    'humidity':daily_resample['humidity'].mean(),
    'rain': daily_resample['rain_1h'].sum(),  # Total rain for the day
    'snow': daily_resample['snow_3h'].sum(),  # Total snow for the day
    'cloud_cover': daily_resample['clouds_all'].mean()  # Average cloud cover
})

Barcelona_data = Barcelona_data.add_prefix('Barcelona_')
```

```python
# Converting Hourly to daily Data:
Seville_full_weather_data.set_index('dt_iso', inplace=True)

# Resample by day and calculate daily statistics
daily_resample = Seville_full_weather_data.resample('D')

# Calculate daily statistics for the new DataFrame
Seville_data = pd.DataFrame({
    'temp_min': daily_resample['temp'].min(),
    'temp_max': daily_resample['temp'].max(),
    'temp_avg': daily_resample['temp'].mean(),
    'pressure': daily_resample['pressure'].mean(),
    'wind_speed': daily_resample['wind_speed'].mean(),
    'humidity':daily_resample['humidity'].mean(),
    'rain': daily_resample['rain_1h'].sum(),  # Total rain for the day
    'snow': daily_resample['snow_3h'].sum(),  # Total snow for the day
    'cloud_cover': daily_resample['clouds_all'].mean()  # Average cloud cover
})

Seville_data = Seville_data.add_prefix('Seville_')
```

```python
# Converting Hourly to daily Data:
Madrid_full_weather_data.set_index('dt_iso', inplace=True)

# Resample by day and calculate daily statistics
daily_resample = Madrid_full_weather_data.resample('D')

# Calculate daily statistics for the new DataFrame
Madrid_data = pd.DataFrame({
    'temp_min': daily_resample['temp'].min(),
    'temp_max': daily_resample['temp'].max(),
    'temp_avg': daily_resample['temp'].mean(),
    'pressure': daily_resample['pressure'].mean(),
    'wind_speed': daily_resample['wind_speed'].mean(),
    'humidity':daily_resample['humidity'].mean(),
    'rain': daily_resample['rain_1h'].sum(),  # Total rain for the day
    'snow': daily_resample['snow_3h'].sum(),  # Total snow for the day
    'cloud_cover': daily_resample['clouds_all'].mean()  # Average cloud cover
})

Madrid_data = Madrid_data.add_prefix('Madrid_')
```

```python
# List of DataFrames to be merged horizontally
dataframes_to_merge = [Valencia_data, Seville_data, Madrid_data, Barcelona_data, Bilbao_data]

# Merge horizontally using concat with axis=1
weather_merged_df = pd.concat(dataframes_to_merge, axis=1)  # Axis 1 indicates horizontal merge
# Display the merged DataFrame
print("Horizontally Merged DataFrame:")
print(weather_merged_df.head()) 
```

```python
# Reset the index to make 'dt_iso' a regular column
weather_merged_df.reset_index(inplace=True)  # Convert index back to a column

# Convert 'dt_iso' to date-only format
weather_merged_df['dt_iso'] = pd.to_datetime(weather_merged_df['dt_iso'], errors='raise').dt.date

# Ensure 'time' is in date-only format in energy_data
energy_data['time'] = pd.to_datetime(energy_data['time'], errors='raise').dt.date

# Merge the DataFrames on 'dt_iso' and 'time'
merged_df = pd.merge(
    weather_merged_df,
    energy_data,
    left_on='dt_iso',  # Use 'dt_iso' from weather_merged_df
    right_on='time',   # Use 'time' from energy_data
    how='inner'  # Use 'inner' join for matching rows
)


# Convert 'dt_iso' to datetime format
merged_df['dt_iso'] = pd.to_datetime(merged_df['dt_iso'], errors='raise')  # Convert to datetime

# Create a new column to determine if it's a weekend or weekday
merged_df['day_of_week'] = merged_df['dt_iso'].dt.dayofweek  # 0=Monday, 6=Sunday

# Define whether it's a weekend or weekday
merged_df['is_weekend'] = merged_df['day_of_week'].apply(lambda x: 'Weekend' if x >= 5 else 'Weekday')

# Create a new column with the month extracted from 'dt_iso'
merged_df['month'] = merged_df['dt_iso'].dt.strftime('%b') 
```

```python
# Calculate the average temperature across all cities
merged_df['avg_city_temp'] = merged_df[['Valencia_temp_avg', 'Seville_temp_avg', 'Madrid_temp_avg', 'Barcelona_temp_avg', 'Bilbao_temp_avg']].mean(axis=1)
```

```python
fig = px.scatter(
    merged_df,
    x='avg_city_temp',  # Average temperature
    y='max_total load actual',  # Max electricity generation
    title='Relationship between Average City Temperature and Max Electricity Generation ',
    labels={'x': 'Average City Temperature (°C)', 'y': 'Max Electricity Generation (MW)'},
    color='is_weekend',
    symbol='is_weekend',
      # Add a trendline to visualize the correlation
)

# Set the background to white and remove grid lines
fig.update_layout(
    paper_bgcolor='white',  # Plot background
    plot_bgcolor='white',   # Figure background
    showlegend=True,        # Display legend
    legend=dict(
        traceorder='reversed',  # Show legend in reverse order
        itemsizing='constant',  # Keep legend item sizes consistent
        bgcolor='white',        # Legend background color
    )
)

# Hide specific legend items (like 'is_weekend')
#fig.update_traces(showlegend=False)  # This disables individual traces in the legend

# Update the x-axis and y-axis settings
fig.update_xaxes(
    showline=True,
    title='Average City Temperature in Kelvin',
    linecolor='black',
    linewidth=2,
    showgrid=False  # Hide x-axis grid lines
)

fig.update_yaxes(
    showline=True,
    title='Max Electricity Generation (GWh)',
    linecolor='black',
    linewidth=2,
    showgrid=False  # Hide y-axis grid lines
)

fig.show()
```

```python
# Define custom color mapping based on month groups
color_map = {
    'Dec': 'blue', 'Jan': 'blue', 'Feb': 'blue',  # Winter months
    'Mar': 'green', 'Apr': 'green', 'May': 'green',  # Spring months
    'Jun': 'orange', 'Jul': 'orange', 'Aug': 'orange',  # Summer months
    'Sep': 'red', 'Oct': 'red', 'Nov': 'red'  # Autumn months
}
# Define season mapping
season_map = {
    'Dec': 'Winter', 'Jan': 'Winter', 'Feb': 'Winter',  # Winter months
    'Mar': 'Spring', 'Apr': 'Spring', 'May': 'Spring',  # Spring months
    'Jun': 'Summer', 'Jul': 'Summer', 'Aug': 'Summer',  # Summer months
    'Sep': 'Autumn', 'Oct': 'Autumn', 'Nov': 'Autumn'  # Autumn months
}

# Create new 'season' column based on the mapping
merged_df['season'] = merged_df['month'].map(season_map)  # Map months to seasons


# Map 'month' to custom colors
merged_df['color'] = merged_df['month'].map(color_map)

fig = px.scatter(
    merged_df,
    x='avg_city_temp',  # Average temperature
    y='max_total load actual',  # Max electricity generation
    title='Relationship between Average City Temperature and Max Electricity Generation',
    labels={'x': 'Average City Temperature (°C)', 'y': 'Max Electricity Generation (MW)'},
    color='season'
    
      # Add a trendline to visualize the correlation
)

# Update the x-axis and y-axis settings
fig.update_xaxes(
    showline=True,
    title='Average City Temperature in Kelvin',
    linecolor='black',
    linewidth=2,
    showgrid=False  # Hide x-axis grid lines
)

fig.update_yaxes(
    showline=True,
    title='Max Electricity Generation (GWh)',
    linecolor='black',
    linewidth=2,
    showgrid=False  # Hide y-axis grid lines
)

fig.update_layout(
    paper_bgcolor='white',  # Plot background
    plot_bgcolor='white',   # Figure background
    showlegend=True,        # Display legend
    legend=dict(
        traceorder='reversed',  # Show legend in reverse order
        itemsizing='constant',  # Keep legend item sizes consistent
        bgcolor='white',        # Legend background color
    )
)
# Display the scatter plot
fig.show()
```

# Electricity Total Demand Prediction:

```python

merged_df_for_analysis = merged_df[['dt_iso', 'Valencia_temp_min', 'Valencia_temp_max', 'Valencia_temp_avg',
       'Valencia_pressure', 'Valencia_wind_speed', 'Valencia_humidity',
       'Valencia_rain', 'Valencia_snow', 'Valencia_cloud_cover',
       'Seville_temp_min', 'Seville_temp_max', 'Seville_temp_avg',
       'Seville_pressure', 'Seville_wind_speed', 'Seville_humidity',
       'Seville_rain', 'Seville_snow', 'Seville_cloud_cover',
       'Madrid_temp_min', 'Madrid_temp_max', 'Madrid_temp_avg',
       'Madrid_pressure', 'Madrid_wind_speed', 'Madrid_humidity',
       'Madrid_rain', 'Madrid_snow', 'Madrid_cloud_cover',
       'Barcelona_temp_min', 'Barcelona_temp_max', 'Barcelona_temp_avg',
       'Barcelona_pressure', 'Barcelona_wind_speed', 'Barcelona_humidity',
       'Barcelona_rain', 'Barcelona_snow', 'Barcelona_cloud_cover',
       'Bilbao_temp_min', 'Bilbao_temp_max', 'Bilbao_temp_avg',
       'Bilbao_pressure', 'Bilbao_wind_speed', 'Bilbao_humidity',
       'Bilbao_rain', 'Bilbao_snow', 'Bilbao_cloud_cover', 'max_total load actual', 
       'is_weekend']]

# Ensure 'dt_iso' is in datetime format
merged_df_for_analysis['dt_iso'] = pd.to_datetime(merged_df_for_analysis['dt_iso'], errors='raise')

# One-hot encode 'is_weekend'
merged_df_for_analysis = pd.get_dummies(merged_df_for_analysis, columns=['is_weekend'], prefix='is_weekend', drop_first=True)
```

```python
features = [ 'Valencia_temp_avg','Valencia_wind_speed', 'Valencia_humidity',
       'Valencia_rain',  'Valencia_cloud_cover','Seville_temp_avg',
    'Seville_wind_speed', 'Seville_humidity','Seville_rain',  'Seville_cloud_cover',
    'Madrid_temp_avg','Madrid_wind_speed', 'Madrid_humidity','Madrid_rain', 'Madrid_cloud_cover',
      'Barcelona_temp_avg','Barcelona_wind_speed', 'Barcelona_humidity',
       'Barcelona_rain',  'Barcelona_cloud_cover','Bilbao_temp_avg',
'Bilbao_wind_speed', 'Bilbao_humidity','Bilbao_rain', 'Bilbao_cloud_cover',  
       'is_weekend_Weekend']

# Define the split point
split_point = '2017-12-31'

# Training data before the split point
train_data = merged_df_for_analysis[merged_df_for_analysis['dt_iso'] <= split_point]

# Testing data after the split point
test_data = merged_df_for_analysis[merged_df_for_analysis['dt_iso'] > split_point]

# Target variable
target = 'max_total load actual'
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import plotly.express as px

# Train the Random Forest model
model = RandomForestRegressor(n_estimators=500, random_state=42)  # Adjust as needed
model.fit(train_data[features], train_data[target])

# Predict on the testing set
y_pred = model.predict(test_data[features])

# Calculate Mean Squared Error to evaluate model performance
mse = mean_squared_error(test_data[target], y_pred)
print("Mean Squared Error:", mse)

```

```python
# Scatter plot of actual vs. predicted values
fig = px.scatter(
    x=test_data['dt_iso'],  # X-axis data (time)
    y=test_data[target],  # Y-axis data (actual values)
    title='Actual vs. Predicted Max Electricity Generation',  # Plot title
)

# Add a second trace for predicted values
fig.add_scatter(
    x=test_data['dt_iso'],  # X-axis data (time)
    y=y_pred,  # Y-axis data (predicted values)
    mode='lines',  # Line mode to connect points
    name='Predicted'  # Trace name for the legend
)

# Update layout to set background color to white and display axes
fig.update_layout(
    xaxis=dict(
        title='Days',  # Set x-axis title
        showline=True,  # Display x-axis line
        linecolor='black',  # Set x-axis line color
        linewidth=2,  # Set x-axis line width
    ),
    yaxis=dict(
        title='Electricity Generation in GWh',  # Set y-axis title
        showline=True,  # Display y-axis line
        linecolor='black',  # Set y-axis line color
        linewidth=2,  # Set y-axis line width
    ),
    paper_bgcolor='white',  # Set plot background color to white
    plot_bgcolor='white',  # Set figure background color to white
)

# Display the updated plot
fig.show()
```
